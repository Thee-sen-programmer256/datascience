```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix,classification_report,mean_absolute_error
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('/Users/_senyonjo/Downloads/teams.csv')
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df
```

```python
df.isna().any(axis=0)
```

```python
df.dropna(inplace=True)
```

```python
df.isna().any(axis=0)
```

```python
#splitting the data for the predictions 
trainset=df[df['year']<2012].copy()
testset=df[df['year']>=2012].copy()
```

```python
#we define our predictors and the traget
target=['medals']
predictors=['events','athletes','prev_medals']

#we create a insstance of the model
model=LinearRegression()

#we fir the model
model.fit(trainset[predictors],trainset[target])
```

```python
#predicting the testset
model.predict(testset[predictors])
```

```python
testset['predictions']=model.predict(testset[predictors])
```

```python
testset
```

```python
testset.loc[testset['predictions']<0,'predictions']=0
```

```python
testset['predictions']=testset['predictions'].round()
```

```python
testset
```

```python
#checking the meanaberr
meanaberr=mean_absolute_error(testset['medals'],testset['predictions'])
```

```python
meanaberr
```

```python
absoluteerror=(testset['medals']-testset['predictions']).abs()
absoluteerror
```

```python
errorperteam=absoluteerror.groupby(testset['team']).mean()
errorperteam
```

```python
medalperteam=testset['medals'].groupby(testset['team']).mean()
medalperteam
```

```python
erroratio=errorperteam/medalperteam
erroratio
```

```python
erroratio=erroratio[~pd.isnull(erroratio)]
erroratio
```

```python
erroratio=erroratio[np.isfinite(erroratio)]
erroratio.sort_values()
```
